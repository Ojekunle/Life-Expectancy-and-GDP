1. DataProcessor Usage Example

The DataProcessor class is used to clean and prepare data by handling missing values and ensuring good data consistency.


```python
# Import the package from Python_Project
from python_project import DataProcessor

# Load a sample dataset
import pandas as pd

# Replace with the path to your actual data file
data = pd.read_csv('/Users/user/Downloads/Life-Expectancy-and-GDP-Starter/all_data.csv')  

# Create an instance of the DataProcessor class
processor = DataProcessor(data)

# Clean the data
cleaned_data = processor.clean_data()

# Display the cleaned data
print("Cleaned Data:")
print(cleaned_data.head())
```

2. DataAnalyzer Usage Example

The DataAnalyzer class performs statistical analysis on the data, such as calculating basic statistics and analyzing life expectancy trends.

```python
# Import DataAnalyzer from Python_Project
from python_project import DataAnalyzer

# Create an instance of DataAnalyzer with the cleaned data
analyzer = DataAnalyzer(cleaned_data)

# Calculate and display basic statistics for life expectancy and GDP
life_stats, gdp_stats = analyzer.calculate_statistics()
print("Life Expectancy Statistics:\n", life_stats)
print("GDP Statistics:\n", gdp_stats)

# Analyze life expectancy trends over the years
average_life_expectancy = analyzer.analyze_life_expectancy()
print("\nAverage Life Expectancy by Year:")
print(average_life_expectancy)
```

3. CorrelationAnalyzer Usage Example

The CorrelationAnalyzer class analyzes and visualizes correlations between numerical variables in the dataset.

```python
# Import CorrelationAnalyzer from Python_Project
from python_project import CorrelationAnalyzer

# Create an instance of CorrelationAnalyzer with the cleaned data
correlation_analyzer = CorrelationAnalyzer(cleaned_data)

# Compute and display the correlation matrix
correlation_matrix = correlation_analyzer.compute_correlation()
print("Correlation Matrix:\n", correlation_matrix)

# Plot the correlation matrix as a heatmap
correlation_analyzer.plot_correlation()
```

4. PredictiveModel Usage Example

The PredictiveModel class is used to train a linear regression model to predict future target values based on feature values.

```python
# Import PredictiveModel from Python_Project
from python_project import PredictiveModel

# Create an instance of PredictiveModel with the cleaned data
predictive_model = PredictiveModel(cleaned_data)

# Train the model using 'GDP' as the feature and 'Life expectancy at birth (years)' as the target
predictive_model.train_model('GDP', 'Life expectancy at birth (years)')

# Predict future life expectancy based on hypothetical future GDP values
future_gdp_values = [8.5e+10, 9.0e+10, 9.5e+10]  # Example GDP values
predicted_life_expectancy = predictive_model.predict_future(future_gdp_values)
print(f"Predicted Future Life Expectancy for GDP values {future_gdp_values}:")
print(predicted_life_expectancy)
```
